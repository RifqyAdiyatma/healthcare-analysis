## Data Cleaning

```python
df.head()
```

```python
df_clean = df.copy()
```

```python
#See if there is null value in each columns
df_clean.isnull().sum()
```

Because there are no missing/null value in columns, next is drop column that's not going to be used

1. Drop Columns

```python
df_clean = df_clean.drop(columns = ['Room Number', 'Doctor'])
```

2. Remove Duplicates

```python
df_clean.drop_duplicates()

df_clean.head()
```

3. Name 

```python
# Create a function to capitalise the name
def capitalize_name(name):
    parts = name.split()
    if len(parts) >= 2:
        first_name = parts[0].capitalize()
        last_name = parts[1].capitalize()
    else:
        first_name = parts[0].capitalize()
        last_name = ''
    return pd.Series([first_name, last_name])
```

```python
df_clean[['First Name', 'Last Name']] = df_clean['Name'].apply(capitalize_name)
df_clean.head()
```

```python
#Move first and last name to the first column
cols = list(df_clean.columns)
df_clean = df_clean[[cols[-1]] + cols[:-1]]
df_clean.head()
```

```python
cols = list(df_clean.columns)
df_clean = df_clean[[cols[-1]] + cols[:-1]]
df_clean.head()
```

```python
#Drop name column
df_clean = df_clean.drop(columns = 'Name')
df_clean.head()
```

4. Blood Type

```python
df_clean.value_counts('Blood Type', dropna = False)
```

```python
df_clean.value_counts('Medical Condition', dropna = False)
```

5. Changing Date type

Date of Admission

```python
df_clean['Date of Admission'] = pd.to_datetime(df_clean['Date of Admission'], errors = 'coerce')

df_clean.head(10)
```

Discharge Date

```python
df_clean['Discharge Date'] = pd.to_datetime(df_clean['Discharge Date'], errors = 'coerce')

df_clean.head(10)
```

6. Length of Stay

Create new column to count how many days a patient stay in hospital

```python
df_clean['Length of Stay'] = (df_clean['Discharge Date'] - df_clean['Date of Admission']).dt.days
df_clean.head()
```

7. Billing Amount

Remove the number after comma 

```python
df_clean['Billing Amount'] = df_clean['Billing Amount'].astype('str').str.split('.')
df_clean.head(10)
```

```python
df_clean['Billing Amount'] = df_clean['Billing Amount'].str[0].astype('int')
df_clean.head(10)
```

8. Admission Type

```python
df_clean.value_counts('Admission Type', dropna = False)
```

There is no error or missing value in this column.

8. Medication

```python
df_clean.value_counts('Medication', dropna = False)
```

9. Test Result

```python
df_clean.value_counts('Test Results', dropna = False)
```

## Exploratory Data Analysis 

### Descriptive Statistics

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
df_condition = df.groupby('Medical Condition', as_index = False).agg(Total = ('Medical Condition', 'count'))
df_condition.sort_values('Total', ascending = False)
```

```python
df_blood = df.groupby('Blood Type', as_index = False).agg(Blood_Count = ('Blood Type', 'count'))
df_blood.sort_values('Blood_Count', ascending = False)
```

### Data Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.hist(bins = 30, figsize = (20, 15))

for column in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(df, y=column)
    plt.show()
```

```python
plt.bar(df_blood['Blood Type'], df_blood['Blood_Count'], color = 'red')

plt.xlabel('Blood Type')
plt.ylabel('Total')
plt.title('T')
```

```python
sns.countplot(data=df, x='Medical Condition', hue = 'Medical Condition', order = df['Medical Condition'].value_counts().index[0:], palette = 'flare')
plt.show()
```

Bar chart visualisation is not really good because the numbers are not much different. So change the visualisation to pie chart

```python
plt.pie(df_blood['Blood_Count'], labels = df_blood['Blood Type'], startangle = 90, 
       autopct = '%1.2f%%', colors = sns.color_palette('flare'))
plt.show()
```

```python
plt.pie(df_condition['Total'], labels = df_condition['Medical Condition'], startangle = 90, 
       autopct = '%1.2f%%', colors = sns.color_palette('flare'))
plt.show()
```

#### Billing Amount

Total Billing Amount by Medical Condition

```python
group_data = df.groupby('Medical Condition')['Billing Amount'].sum().reset_index()

group_data = group_data.sort_values('Billing Amount', ascending = False)

group_data.head()
```

```python
sns.barplot(data=group_data, x='Billing Amount', y='Medical Condition', errorbar=None, color = 'red')
plt.title('Total Billing Amount by Medical Condition')
plt.xlabel('Total Billing Amount')
plt.ylabel('Medical Condition')
plt.show()
```

Average Billing Amount by Insurance Provider

```python
insurance = df.groupby('Insurance Provider')['Billing Amount'].mean().reset_index()

insurance = insurance.sort_values('Billing Amount', ascending = False)

insurance
```

```python
sns.barplot(data=insurance, x='Billing Amount', y='Insurance Provider', errorbar=None, color = 'red')
plt.title('Average Billing Amount by Insurance Provider')
plt.xlabel('Average Billing Amount')
plt.ylabel('Insurance Provider')
plt.show()
```

Total Billing Amount by Hospital

```python
hospital = df.groupby('Hospital')['Billing Amount'].sum().reset_index()

hospital = hospital.sort_values('Billing Amount', ascending = False)[hospital['Billing Amount'] > 800000]

hospital
```

```python
plt.figure(figsize = (15, 10))
sns.barplot(data=hospital, x='Hospital', y='Billing Amount', errorbar=None, color = 'red')
plt.title('Total Billing Amount by Insurance Provider')
plt.xlabel('Total Billing Amount')
plt.ylabel('Insurance Provider')
plt.show()
```
